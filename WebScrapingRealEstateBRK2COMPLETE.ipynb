```python
house_locations = []
house_prices = []
# house_features = []
house_bedrooms = []
house_bathrooms = []
# keys = ['Bedrooms', 'Bathrooms']


for j in range(50):
    parent_div = driver.find_elements(By.XPATH, '//div[@class="flex flex-wrap gap-x-2 gap-y-1"]')
    house_location = driver.find_elements(By.XPATH, '//div[@class="flex max-w-full items-center"]/p[@class="ml-1 truncate text-sm font-normal capitalize text-grey-650"]')
#     for i in range(len(house_location)):
#         house_locations.append(house_location[i].text)
    for location in house_location:
        house_locations.append(location.text)
    house_price = driver.find_elements(By.XPATH, '//div[@class="relative w-full overflow-hidden rounded-2xl bg-white"]')#.get_attribute("data-bi-listing-price")
    for price in house_price:
#         if price.text != '':
        house_prices.append(price.get_attribute("data-bi-listing-price")) 
#     features = driver.find_elements(By.XPATH, '//*[@class="whitespace-nowrap font-normal"]')
#     for feature in features:
#         house_bedrooms.append(feature.get_attribute)
    for i in range(len(parent_div)):
#     name= parent_div[0].find_element(By.XPATH, f'(//span[@data-cy="card-bathrooms"])[{i}]')
        try:
            name= parent_div[0].find_element(By.XPATH, f'(//span[@data-cy="card-bathrooms"])[{i+1}]')
#             if name.text == '':
#                 house_bathrooms.append('0')
#             else:
            house_bathrooms.append(name.text)
        except NoSuchElementException:
            house_bathrooms.append(None)
    
    bedrooms = driver.find_elements(By.XPATH, '//span[@data-cy="card-beds"]')
    for bedroom in bedrooms:
        house_bedrooms.append(bedroom.text)
        
#     house_prices.append('______________________________________')
#     house_locations.append('______________________________________')
#     house_bedrooms.append('______________________________________')
#     house_bathrooms.append('______________________________________')
    
#     bathrooms = driver.find_elements(By.XPATH, '//span[@data-cy="card-bathrooms"]')
#     for bathroom in bathrooms:
#         house_bathrooms.append(bathroom.text)

            
    
    
#     for i in range(0, len(features), 2):
#         prop_features.append({keys[0]: features[i].text, keys[1]: features[i + 1].text})
    
    next_button = driver.find_element(By.XPATH, '//*[@class="fill-current transform -rotate-90 inline-block text-secondary w-3"]')
    next_button.click()
    time.sleep(3)
```
